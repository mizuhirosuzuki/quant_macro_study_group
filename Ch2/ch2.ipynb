```julia
θ0_grid = -0.001:0.00001:0.001;
θ1_grid = 0.35:0.0001:0.36;
resid_array = zeros(length(θ0_grid), length(θ1_grid));
```

```julia
for (i, θ0) in enumerate(θ0_grid), (j, θ1) in enumerate(θ1_grid)
    resid_array[i, j] = sum(resid_projection([θ0, θ1], params).^2)
end
```

```julia
heatmap(θ1_grid, θ0_grid, log.(resid_array), c = cgrad([:orange, :blue]))
xlabel!("θ1")
ylabel!("θ0")
```
