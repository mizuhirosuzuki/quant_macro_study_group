```julia
g(x) = sin(x[1] + cos(x[2]))
lb, ub = [1,3], [2, 4] # lower and upper bounds of the domain, respectively
x = chebpoints((10,20), lb, ub)
c = chebinterp(g.(x), lb, ub)
```

```julia
function chebyshev_approxfun(x_grid)
    S = Chebyshev(-1..1);
    collocation_points = points(S, 50);
    y_evals = exp.(- collocation_points);
    fapprox = Fun(S, ApproxFun.transform(S, y_evals));
    return fapprox.(x_grid)
end
```

```julia
@benchmark chebyshev_approxfun(x)
```

```julia
function chebyshev_compecon(x)
    basis = fundefn(:cheb, 50, -1, 1)
    c = funfitf(basis, y -> exp.(-y))
    return funeval(c, basis, x)
end
```

```julia
g(x) = sin(x[1] + cos(x[2]))
lb, ub = [1,3], [2, 4] # lower and upper bounds of the domain, respectively
x = chebpoints((10,20), lb, ub)
c = chebinterp(g.(x), lb, ub)
```

```julia
function chebshev_2d_fast1(x)
    g(x) = sin(x[1] + cos(x[2]))
    lb, ub = [1,3], [2, 4] # lower and upper bounds of the domain, respectively
    xx = chebpoints((10,20), lb, ub)
    c = chebinterp(g.(xx), lb, ub)
    return c.(x)
end
```

```julia
x3 = [[i, j] for i in 1:0.1:2 for j in 3:0.1:4];
```

```julia
@benchmark chebshev_2d_fast1(x3)
```

```julia
function chebshev_2d_compecon(x)
    g(z) = sin(z[1] + cos(z[2]))

    basis = fundefn(:cheb, (10, 20), (1, 3), (2, 4))
    c = funfitf(basis, z -> mapslices(g, z, dims=2));
    return map(z -> funeval(c, basis, z)[1][1], x)
end
```

```julia
@benchmark chebshev_2d_compecon(x3)
```
